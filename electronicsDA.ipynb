```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

# Reading Data

```python
df=pd.read_csv('mobile_sales_data.csv',parse_dates=['Inward Date','Dispatch Date'])
```

```python
mob_phone=df[df['Product']=='Mobile Phone']
```

```python
laptop=df[df['Product']=='Laptop']
laptop
```

# Mobile Analysis

```python
mob_phone
```

```python
total_sales_mob=mob_phone.groupby('Brand')['Price'].sum().reset_index()
total_sales_mob.sort_values(by='Price',ascending=False)
```

Total Sales

```python
plt.figure(figsize=(15,10))
sns.barplot(data=total_sales_mob,x='Brand',y='Price')
plt.xlabel('Mobile phone Brands')
plt.show()
```

Sales By region

```python
region_sales=mob_phone.groupby('Region')['Price'].sum().reset_index()
region_sales
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data=region_sales,x='Region',y='Price')
plt.xlabel("Regions Mobile sales")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
mob_phone['Dispatch_month']=mob_phone['Dispatch Date'].dt.to_period('M').astype(str)
```

```python
SumOfPrice=mob_phone.groupby(['Dispatch_month'])['Price'].sum().reset_index()
SumOfPrice.sort_values(by='Dispatch_month',ascending=False)
SumOfPrice
```

## Mobile Sales Trend


```python
plt.figure(figsize=(20,20))
x=sns.lineplot(data=SumOfPrice,x='Dispatch_month',y='Price')
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=90)
plt.show()

```

## Total Sales For Brand

```python
plt.figure(figsize=(20,20))
x=sns.barplot(data=total_sales_mob,x='Brand',y='Price')
plt.xlabel("Mobile phone Brands")
plt.ylabel("Price")
for contanier in x.containers:
    x.bar_label(contanier)
plt.show()
```

```python
mob_phone
```

```python
quantity_sold_mob=mob_phone.groupby('Brand')['Quantity Sold'].sum().reset_index()
quantity_sold_mob
```

### Mobile Brands Total quntity Sales

```python
plt.figure(figsize=(20,20))
x=sns.barplot(data=quantity_sold_mob,x='Brand',y='Quantity Sold')
plt.xlabel('Brands')
plt.ylabel('Total Sold items')
for contanier in x.containers:
    x.bar_label(contanier)
plt.show()
```

```python
saleslist=[Sales_Processor_specs,sales_Ram,sales_Rom,sales_SSD]
```

```python
for x in saleslist:
    plt.figure(figsize=(12,12))
    X=sns.barplot(data=x,x=x.columns[0],y=x.columns[1])
    for contanier in X.containers:
        X.bar_label(contanier)
    plt.show()
```
